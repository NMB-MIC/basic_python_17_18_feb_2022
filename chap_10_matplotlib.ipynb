## Multiple figures and update figures

```python
x = np.linspace(-1.4,1.4,30)

plt.figure(1)
plt.subplot(211)
plt.title("Square and Cube")
plt.plot(x,x**2)
plt.subplot(212)
plt.plot(x,x**3)

plt.figure(2)
plt.subplot(121)
plt.plot(x,x**4)
plt.title("y = x**4")
plt.subplot(122)
plt.plot(x,x**5)
plt.title("y = x**5")

#update figure
plt.figure(1)
plt.plot(x,-x**3,"r:")

plt.show()
```

## Drawing text

```python
x = np.linspace(-1.5,1.5,30)

px = 0.8
py = px**2
plt.plot(x,x**2,"b-",px,py,"ro")

plt.text(0,1.5,"Squre function\n $y = x^2$",fontsize=20,color="green",horizontalalignment="center")
plt.text(px-0.08,py,"Bestest point", ha="right" , weight="heavy")
plt.text(px,py, "x = %0.2f\ny= %0.2f"%(px,py) , rotation=30 ,color="gray")
plt.show()
```

```python
# anotate
plt.plot(x,x**2,px,py,"ro")
plt.annotate("Bestest point",
                xy=(px,py),
                xytext=(px-1.3,py+0.5),
                weight="heavy",
                fontsize=14,
                arrowprops={"facecolor":"lightgreen"})
plt.show()
```

```python
#bbox
plt.plot(x,x**2,px,py,"ro")

bbox_props = dict(boxstyle="rarrow,pad=0.3",ec="green",lw=2,fc="lightblue")
plt.text(px-0.2,py,"Bestest point",bbox=bbox_props, ha="right")

bbox_props = dict(boxstyle="round4,pad=1,rounding_size=0.1",ec="orange",fc="lightgreen",lw=5)
plt.text(0,1.5,"Square function\n$y = x^2$",bbox=bbox_props,ha="center",fontsize=20,color="blue")
plt.show()
```

## Polar projection

```python
radius = 1
theta = np.linspace(0,2*np.pi*radius,1000)

plt.subplot(111,projection="polar")
plt.plot(theta,np.sin(5*theta),"g-")
plt.plot(theta,0.5*np.cos(20*theta),"b-")
plt.show()
```

## 3D projection

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)
X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

figure = plt.figure(1, figsize= (12,4))
subplot3d = plt.subplot(111,projection="3d")
surface = subplot3d.plot_surface(X,Y,Z, rstride=1,cstride=1,cmap=matplotlib.cm.coolwarm,linewidth=0.1)
plt.show()
```

```python
#display this same data is via a contour plot
plt.contourf(X,Y,Z,cmap=matplotlib.cm.coolwarm)
plt.colorbar()
plt.show()
```

## Scatter plot

```python
from numpy.random import rand
x,y = rand(2,100)

plt.scatter(x,y)
plt.show()
```

```python
#scalable of each point
x,y,scale = rand(3,100)
scale = 500* scale **5

plt.scatter(x,y,s=scale)
plt.show()
```

```python
#edge colors and the alpha level

for color in ['red','green','blue']:
    n = 100
    x,y = rand(2,n)
    scale = 500* rand(n) ** 5
    plt.scatter(x,y, s=scale, c=color,alpha=0.3, edgecolors="blue")

plt.show()
```

## Lines

```python
from numpy.random import randn

def plot_line(axis,slope,intercept,**kargs):
    xmin , xmax = axis.get_xlim()
    plt.plot([xmin,xmax],[xmin*slope+intercept , xmax*slope+intercept],**kargs)

x = randn(1000)
y = 0.5*x + 5 + randn(1000)*2
plt.axis([-2.5,2.5,-5,15])
plt.scatter(x,y,alpha=0.2)
plt.plot(1,0,"ro")
plt.vlines(1,-5,0,color="red")
plt.hlines(0,2.5,1,color="red")
plot_line(axis=plt.gca(),slope=0.5,intercept=5,color="magenta")
plt.grid(True)
plt.show()
```

## Animations

```python
import matplotlib.animation as animation
```

```python
#create data set
x = np.linspace(-1,1,100)
y = np.sin(x**2*25)
data = np.array([x,y])

#defualt
fig = plt.figure()
line, = plt.plot([],[],"r-")
plt.axis([-1.1,1.1,-1.1,1.1])
plt.plot([-0.5,0.5],[0,0],"b-",[0,0],[-0.5,0.5],"b-",0.0,"ro")
plt.grid(True)

def update_line(num,data,line):
    line.set_data(data[..., :num] + np.random.rand(2,num)/25)
    return line
    
```

```python
animation.FuncAnimation(fig,update_line,frames=50,fargs=(data,line),interval=100)
```
